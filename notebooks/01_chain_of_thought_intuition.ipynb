# 01 — Chain-of-thought intuition

The simplest reasoning trick: tell the model to *think out loud*.

We'll use plain `gpt2` (124M, CPU-fine) so the trick is the only variable.

```python
from transformers import pipeline, set_seed
set_seed(0)
gen = pipeline('text-generation', model='gpt2', device=-1)
```

## Direct prompt

```python
prompt_direct = 'Q: If I have 3 apples and buy 2 more, how many apples do I have?\nA:'
print(gen(prompt_direct, max_new_tokens=20, do_sample=False)[0]['generated_text'])
```

## Chain-of-thought prompt

```python
prompt_cot = 'Q: If I have 3 apples and buy 2 more, how many apples do I have?\nA: Let\u2019s think step by step.'
print(gen(prompt_cot, max_new_tokens=60, do_sample=False)[0]['generated_text'])
```

## Takeaway

GPT-2 isn't a strong reasoner — but you can already see the *shape* of the trick: giving the model token budget to lay out intermediate steps changes its output distribution.

**Real reasoning models bake this trick into the weights** — that's what Module 3 is about.
